```python
#Función de entrenamiento

def entrenar(x, y, EPOCHS, imprimir_error_cada, lr):
    
    observacion = np.matrix([x, np.ones_like(x)])
    observacion = observacion.transpose()
    
    m = 10
    b = -5
    parametro = np.array([m, b])
    
    modelo = np.empty((0, 2))
    errores = np.array([])
    
    for i in range(1, EPOCHS + 1):
        #Calculando yhat = mx + b
        yhat = np.matmul(observacion, parametro)
        yhat = np.asarray(yhat).reshape(-1)
   
        #Guardando modelo
        modelo = np.append(modelo, np.array([[m, b]]), axis = 0)
        
        #Calculando error
        error = (1 / 2) * np.nanmean(np.power((y - yhat), 2))    
        errores = np.append(errores, error)

        #Calculando gradientes
        
        gradiente_m = np.mean((yhat - y) * x)
        gradiente_b = np.mean((yhat - y))
        
        #Actualizando parámetros
        m -= lr * gradiente_m
        b -= lr * gradiente_b
        parametro = np.array([m, b])

        if i % imprimir_error_cada != 0:
            continue
        #print('Iteración', i, ': Error =', error, 'y =', np.mean(y), 'yhat =', np.mean(yhat))
        print('Error en iteración', i, ':', error)
             
    print('---------------------------')
    #print('Resultado modelo: Error =', error, 'y =', np.mean(y), 'yhat =', np.mean(yhat))
    print('Error final: ', error)
    
    return modelo, errores
```

```python
#Función para graficar error

def graficarError(errores):
    plt.plot(np.arange(1, len(errores) + 1), errores)
    plt.xlabel('Iteration')
    plt.ylabel('Error')
    plt.show()
    
```

```python
#Función para graficar entrenamiento de modelo

def graficarModelo(modelo, x, y, n):
    
    for i in range(1, len(modelo) + 1):
        if i % n != 0:
            continue
        
        m = modelo[i - 1, 0]
        b = modelo[i - 1, 1]
        yhat = m * x + b
        plt.scatter(x, y)
        plt.plot(x, yhat, color='red')
        plt.title('Iteration ' + str(i))
        plt.xlabel('x')
        plt.ylabel('y')
        plt.show()
    
```

```python
#Función para entrenamiento de regresión lineal con scikit-learn

def entrenarSKL(x, y):
    regr = linear_model.LinearRegression()
    x = x.reshape(-1, 1)
    y = y.reshape(-1, 1)
    regr.fit(x, y)
    modelo = [regr.coef_[0][0], regr.intercept_[0]]
    y_pred = regr.predict(x)
    errores = mean_squared_error(y, y_pred)
    return modelo, errores
```

```python
#Función para comparar dos modelos

def compararModelos(modelo1, modelo2, x):
    m1 = modelo1[0]
    b1 = modelo1[1]
    y1_pred = m1 * x + b1
    
    m2 = modelo2[0]
    b2 = modelo2[1]
    y2_pred = m2 * x + b2
    
    y_promedio = (y1_pred + y2_pred) / 2
    
    return y1_pred, y2_pred, y_promedio
    
```

```python
#Comparación de modelos para X1
modelo1_X1 = modeloX1[-1, 0:2]
modelo2_X1, erroresModelo2_X1 = entrenarSKL(x1, y)
compararModelos(modelo1_X1, modelo2_X1, x1)
```

```python
#Comparación de modelos para X2
modelo1_X2 = modeloX2[-1, 0:2]
modelo2_X2, erroresModelo2_X2 = entrenarSKL(x2, y)
compararModelos(modelo1_X2, modelo2_X2, x2)
```

```python
#Función para prueba de modelos

def pruebaModelo(modelo, x, y):
    m = modelo[0]
    b = modelo[1]
    y_pred = m * x + b
    
    #Calculando error
    errores = (1 / 2) * np.power((y - y_pred), 2)
    return y_pred, errores[-1:][0]
```

```python
#Obteniendo datos de prueba
y_prueba = datosPrueba[:,0]
x1_prueba = datosPrueba[:,1]
x2_prueba = datosPrueba[:,2]

#Probando modelo manual para X1 y X2
y_prueba1_X1, error_prueba1_X1 = pruebaModelo(modelo1_X1, x1_prueba, y_prueba)
y_prueba1_X2, error_prueba1_X2 = pruebaModelo(modelo1_X2, x2_prueba, y_prueba)
print('Modelo manual')
print('-------------')
print('Error para X1:', error_prueba1_X1)
print('Error para X2:', error_prueba1_X2)

#Probando modelo en SKL para X1 y X2
y_prueba2_X1, error_prueba2_X1 = pruebaModelo(modelo2_X1, x1_prueba, y_prueba)
y_prueba2_X2, error_prueba2_X2 = pruebaModelo(modelo2_X2, x2_prueba, y_prueba)

print('\nModelo SKL')
print('----------')
print('Error para X1:', error_prueba2_X1)
print('Error para X2:', error_prueba2_X2)

#Graficando errores
barraAncho = 0.4
barras1 = [error_prueba1_X1, error_prueba1_X2]
barras2 = [error_prueba2_X1, error_prueba2_X2]

r1 = np.arange(len(barras1))
r2 = [z + barraAncho for z in r1]

plt.bar(r1, barras1, color = 'blue', width = barraAncho, label = "Manual model")
plt.bar(r2, barras2, color = 'pink', width = barraAncho, label = "SKL model")

plt.xticks([r + barraAncho / 2 for r in range(len(barras1))], ['X1', 'X2'])

plt.legend()
plt.show()
```

## Conclusión

El modelo manual presenta un menor nivel de error que el modelo calculado con SKL para las dos variables analizadas, por lo cual es el que se recomienda para ser utilizado como modelo de predicción.
